```python
import pandas as pd
import numpy as np
import pickle
```

```python
model_diabetes = pickle.load(open("../models/diabetes_model.pkl", "rb"))
model_heart = pickle.load(open("../models/heart_model.pkl", "rb"))

print("Models loaded successfully")
```

```python
diabetes_columns = pickle.load(open("../models/diabetes_columns.pkl", "rb"))
heart_columns = pickle.load(open("../models/heart_columns.pkl", "rb"))
```

```python
def predict_diseases(diabetes_input, heart_input):

    # ---- Diabetes ----
    df_diabetes = pd.DataFrame([diabetes_input], columns=diabetes_columns)
    diabetes_result = model_diabetes.predict(df_diabetes)[0]

    # ---- Heart ----
    heart_dict = dict(zip(
        ['age','sex','cp','trestbps','chol','fbs','restecg',
         'thalach','exang','oldpeak','slope','ca','thal'],
        heart_input
    ))

    df_heart = pd.DataFrame([heart_dict])
    df_heart = pd.get_dummies(df_heart, drop_first=True)
    df_heart = df_heart.reindex(columns=heart_columns, fill_value=0)

    heart_result = model_heart.predict(df_heart)[0]

    return diabetes_result, heart_result
```

```python
diabetes_input = [0,16,74,0,0,25.6,0.201,20]
heart_input = [63,1,3,145,233,1,0,150,0,2.3,0,0,1]

d_result, h_result = predict_diseases(diabetes_input, heart_input)

print("Diabetes:", "Positive" if d_result==1 else "Negative")
print("Heart Disease:", "Positive" if h_result==1 else "Negative")
```

```python
def get_diabetes_input():
    print("\nEnter Diabetes Details:")
    pregnancies = int(input("Pregnancies: "))
    glucose = int(input("Glucose Level: "))
    bp = int(input("Blood Pressure: "))
    skin = int(input("Skin Thickness: "))
    insulin = int(input("Insulin: "))
    bmi = float(input("BMI: "))
    dpf = float(input("Diabetes Pedigree Function: "))
    age = int(input("Age: "))

    return [pregnancies, glucose, bp, skin, insulin, bmi, dpf, age]
```

```python
def get_heart_input():
    print("\nEnter Heart Disease Details:")
    age = int(input("Age: "))
    sex = int(input("Sex (1 = Male, 0 = Female): "))
    cp = int(input("Chest Pain Type (0–3): "))
    trestbps = int(input("Resting Blood Pressure: "))
    chol = int(input("Cholesterol: "))
    fbs = int(input("Fasting Blood Sugar (1 = True, 0 = False): "))
    restecg = int(input("Resting ECG (0–2): "))
    thalach = int(input("Max Heart Rate: "))
    exang = int(input("Exercise Induced Angina (1 = Yes, 0 = No): "))
    oldpeak = float(input("Oldpeak: "))
    slope = int(input("Slope (0–2): "))
    ca = int(input("CA (0–4): "))
    thal = int(input("Thal (0–3): "))

    return [age, sex, cp, trestbps, chol, fbs, restecg,
            thalach, exang, oldpeak, slope, ca, thal]
```

```python
diabetes_input = get_diabetes_input()
heart_input = get_heart_input()

d_result = predict_diseases(diabetes_input)

print("\n--- Prediction Result ---")
print("Diabetes:", "Positive" if d_result==1 else "Negative")
print("Heart Disease:", "Positive" if h_result==1 else "Negative")
```

```python
heart_input = get_heart_input()
h_result = predict_diseases(heart_input)
print("Heart Disease:", "Positive" if h_result==1 else "Negative")
```
